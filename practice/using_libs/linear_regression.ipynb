```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.datasets as dataset
import matplotlib.pyplot as graph
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
def to_df(function_name):
    result = function_name()
    return pd.DataFrame(data=result.data, columns=result.feature_names)
```

```python
df = to_df(dataset.load_diabetes)
```

```python
df = df[['age', 'bp']]

sns.lmplot(x='age', y='bp', data=df, order=2)
graph.show()
```

```python
x = np.array(df['age'])
y = np.array(df['bp'])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
x_test = x_test.reshape(-1, 1)
x_train = x_train.reshape(-1,1)
```

```python
regression = LinearRegression()
regression.fit(x.reshape(-1, 1), y)
```

```python
y_pred = regression.predict(x_test)

graph.scatter(x_test, y_test, color='orange')
graph.plot(x_test, y_pred)
graph.show()
```
