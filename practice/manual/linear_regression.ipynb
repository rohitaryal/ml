```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as dataset
```

```python
MAX_ITERATION = 10000
LEARNING_RATE = 0.1
MIN_COST_DIFFERENCE = 1e-5
```

```python
data = dataset.load_diabetes()

feature = np.array(data.data)
target = np.array(data.target)

cost = [float('inf')]
theta = np.zeros(feature.shape[1] + 1)
feature = np.c_[np.ones(feature.shape[0]), feature]
```

```python
for i in range(MAX_ITERATION):
    prediction = feature @ theta
    error = prediction - target
    gradient = (feature.T @ error) / len(target)
    theta -= gradient * LEARNING_RATE

    curr_cost = np.sum(error.T @ error) / (2 * len(target))
    cost.append(curr_cost)

    if abs(cost[-1] - cost[-2]) <= MIN_COST_DIFFERENCE:
        break
```

```python
plt.plot(np.linspace(0, cost[-1], len(cost)), cost, label="Cost")

plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost v/s Iteration")

plt.legend()

plt.show()
```
